```python
#Daniela Vargas A01737287
#Se importan las librerías
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Cargar archivo. Directamente lo cargué en el collab en la carpeta y convertirlo en data frame
data= pd.read_csv("Ventas_totales.csv")
data
```

```python
#identificar valores nulos por columna
valores_nulos=data.isnull().sum()
valores_nulos
```

# **Métodos de Sustitución de valores nulos**

```python
#Realizamos una copia del dataframe
data1=data.copy()
```

```python
#Primer métodod de susttitución de valores nulos
#sustituir valores nulos con promedio o media
data1["salon_ventas"]=data1["salon_ventas"].fillna(round(data["salon_ventas"].mean(),1))#Unicamente valores cuantitativos
```

```python
#corroboramos valores nulos
valores_nulos=data1.isnull().sum()
valores_nulos
```

```python
#Método 2
data1["tarjetas_credito"]=data1["tarjetas_credito"].fillna(round(data["tarjetas_credito"].median(),1))#aplicamo mediana
data1
```

```python
#corroboramos valores nulos
valores_nulos=data1.isnull().sum()
valores_nulos
```

```python
#Tercer método
#Ssutituir valores nulos por un valor numérico en concreto
data1["otros_medios"]=data1["otros_medios"].fillna(10)
data1
```
